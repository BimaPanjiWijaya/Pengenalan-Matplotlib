```python
#library
import matplotlib.pyplot as plt
```

```python
# Membuat plot sederhana dengan cara explicit
x = [4,7,2,8,10]
y = [1,2,9,8,5]
# Membuat objek figure
fig = plt.figure() 
# Membuat objek axes
ax = fig.add_subplot()
# Menambahkan data ke objek axes
ax.plot(x, y)
# Menampilkan visualisasi data
plt.show()
```

```python
# Plotting multiple data dalam satu Axes
import numpy as np
X = np.linspace(10, 50, 90)
y1 = np.cos(X*3)
y2 = np.cos(X)
y3 = np.sin(X/2)
y4 = np.sin(X)
fig, ax = plt.subplots(figsize=(5,2))
ax.plot(X, y1, color='yellow')
ax.plot(X, y2, color='black')
ax.plot(X, y3, color='green') 
ax.plot(X, y4, color='red') 
plt.show()
```

```python
#mengatur style tableau-colorblind10
plt.style.use('tableau-colorblind10')
X = np.linspace(10, 50, 90)
y1 = np.cos(X*3)
y2 = np.cos(X)
y3 = np.sin(X/2)
y4 = np.sin(X)
fig, ax = plt.subplots(figsize=(5,2))
ax.plot(X, y1)
ax.plot(X, y2)
ax.plot(X, y3) 
ax.plot(X, y4) 
plt.show()
```

```python
# Mengatur style ggplot
plt.style.use('ggplot')
X = np.linspace(10, 50, 90)
y1 = np.cos(X*2)
y2 = np.cos(X)
y3 = np.sin(X/3)
y4 = np.sin(X)
fig, ax = plt.subplots(figsize=(5,2))
ax.plot(X, y1)
ax.plot(X, y2)
ax.plot(X, y3) 
ax.plot(X, y4) 
plt.show()
```
